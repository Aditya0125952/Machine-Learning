```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import *
```

```python
df=pd.read_csv('/content/movie_reviews_updated.csv')
df.head()
```

```python
#now split and train the model
x=df[['Review']]
y=df['Sentiment']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#now convert the text into the numbers cuz it will take numbers
vector=TfidfVectorizer(ngram_range=(1,2))
new_x_train=vector.fit_transform(x_train['Review'])
new_x_test=vector.transform(x_test['Review'])
```

```python
#now train the model
model=RandomForestClassifier(n_estimators=100)
model.fit(new_x_train,y_train)
```

```python
score=accuracy_score(y_test,model.predict(new_x_test))
matrix=confusion_matrix(y_test,model.predict(new_x_test))
report=classification_report(y_test,model.predict(new_x_test))
print(score)
print(matrix)
print(report)
```
